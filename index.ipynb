### Config

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import statistics
from scipy import stats
from collections import Counter
```

```python
df = pd.read_csv('student_data.csv')
arrId = []
studentID = 1
for i in range(len(df.index)):
    arrId.append(studentID)
    studentID += 1
df['id'] = arrId
df.head()
```

## Vars Categoricas

```python
cat_vars = [] #criando uma lista em branco
for coluna in list(df.columns): #para colunas do meu DF, farei o seguinte:
  if df[coluna].dtype == 'object': #se for do tipo texto
    print(f'{coluna}: possui {len(df[coluna].unique())} valores unicos') #dou um print para ver quantas valores distintos nós temos
    if len(df[coluna].unique()) <= 10: #Se for até 10 valores distintos
      cat_vars.append(coluna) #Entao eu ja sei que dá para analisar com mais facilidade
```

### quantidade de valores

```python
print('-=-=-= QUANTIDADE DE VALORES =-=-=-')
for i in cat_vars:
    # print(i,':',end=' ')
    # x = 0
    # for j in df[i].unique():
    #     if(x == len(df[i].value_counts())-1):
    #         print(j,'(',df[i].value_counts().values[x],')', end=' ')
    #     else:
    #         print(j,'(',df[i].value_counts().values[x],')', end=', ')
    #     x+=1
    display(df[i].value_counts())
```

### Tabela Frequencia

```python
for i in cat_vars:
    sample = df.groupby(i)['id'].agg(['count'])
    sample['Frequência Relativa'] = sample['count']/sample['count'].sum()
    sample['Frequência Relativa %'] = round(sample['Frequência Relativa']*100)
    display(sample)
```

## Vars Quantitativas

```python
num_vars = [] #criando uma lista em branco
for coluna in list(df.columns): #para colunas do meu DF, farei o seguinte:
  if(coluna != 'id'):
    if df[coluna].dtype == 'int64': #se for do tipo texto
      print(f'{coluna}: possui {len(df[coluna].unique())} valores unicos') #dou um print para ver quantas valores distintos nós temos
      num_vars.append(coluna)
```

#### Medidas de Posicao

- media = a media é obtida a partir da soma das
observações, dividindo-se pelo total de observações
- mmediana = é a observação que ocupa a posição central de
um conjunto de observações. 50% das observações são
menores que ela e 50% das observações são maiores que
ela.
- moda = é a observação que ocupa a posição central de
um conjunto de observações. 50% das observações são
menores que ela e 50% das observações são maiores que
ela.

```python
colNum_analisar = [] #criando uma lista em branco
for coluna in list(df.columns): #para colunas do meu DF, farei o seguinte:
  if(coluna != 'id'):
    if df[coluna].dtype == 'int64' or df[coluna].dtype == 'float64': #se for do tipo int ou float
        colNum_analisar.append((coluna,str(df[coluna].dtype))) #Entao eu ja sei que dá para analisar com mais facilidade

print(colNum_analisar)
```

##### Moda

```python
for tupla in colNum_analisar:
  if tupla[1] == 'int64':
    moda = stats.mode(df[tupla[0]])
    print(f'Para a coluna {tupla[0]}, a moda é {moda}')
```

##### Media

```python
for tupla in colNum_analisar:
  media = df[tupla[0]].mean()
  print(f'Para a coluna {tupla[0]}, a media é {media:.2f}')
```

##### Mediana

```python
for tupla in colNum_analisar:
  mediana = statistics.median(df[tupla[0]])
  print(f'Para a coluna {tupla[0]}, a mediana é {mediana:.2f}')
```
